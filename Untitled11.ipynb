```python
print("\n--- 正在為資料集中的歌曲生成嵌入 (這可能需要一些時間)... ---")
start_time = time.time()

# 為了管理 API 呼叫和運行時間，我們限制生成嵌入的數量
# 實際應用中，您會為所有數據生成嵌入
texts_for_embedding = df_processed['Cleaned_Lyrics'].tolist()
num_to_embed = min(len(texts_for_embedding), 500) # 限制為最多 500 首歌曲的嵌入

all_song_embeddings = []
# Gemini 的 embed_content API 可以處理批次請求，提高效率
# 對於非常大的資料集，您可能需要將 texts_for_embedding_subset 進一步分塊
try:
    response = genai.embed_content(
        model=embedding_model,
        contents=texts_for_embedding[:num_to_embed],
        task_type="SEMANTIC_SIMILARITY"
    )
    all_song_embeddings = [np.array(e.values) for e in response.embeddings]
    print(f"成功生成了 {len(all_song_embeddings)} 首歌曲的嵌入。")

except Exception as e:
    print(f"生成歌曲嵌入時發生錯誤: {e}")
    all_song_embeddings = []

end_time = time.time()
print(f"嵌入生成耗時: {end_time - start_time:.2f} 秒")

if not all_song_embeddings:
    print("錯誤: 無法生成歌曲嵌入，無法進行推薦。請檢查 API 金鑰和網路連線。")
    exit()

# 將嵌入儲存到 DataFrame 中
# 確保 df_processed 的長度與成功生成的嵌入數量匹配
df_processed_with_embeddings = df_processed.head(len(all_song_embeddings)).copy()
df_processed_with_embeddings['embedding'] = list(all_song_embeddings)

# 將所有歌曲嵌入轉換為 NumPy 陣列，方便進行相似度計算
song_embeddings_matrix = np.array(df_processed_with_embeddings['embedding'].tolist())

print(f"所有歌曲嵌入的 NumPy 陣列形狀: {song_embeddings_matrix.shape}")
```

```python
def get_recommendations_from_text(user_input_text, dataframe_with_embeddings, embeddings_matrix, top_n=5):
    """
    根據使用者輸入的文字，推薦最相似的歌曲。

    Args:
        user_input_text (str): 使用者輸入的文字描述。
        dataframe_with_embeddings (pd.DataFrame): 包含 'Song Name', 'Artist', 'embedding' 的 DataFrame。
        embeddings_matrix (np.array): 所有歌曲嵌入的 NumPy 陣列。
        top_n (int): 要推薦的歌曲數量。

    Returns:
        pd.DataFrame: 推薦的歌曲列表，包含 'Song Name', 'Artist', 'Similarity_Score'。
                      如果輸入無效或無推薦，則返回空 DataFrame。
    """
    if not user_input_text.strip():
        print("使用者輸入不能為空。")
        return pd.DataFrame()

    print(f"\n處理使用者輸入: '{user_input_text}'...")

    # 1. 預處理使用者輸入
    cleaned_user_input = preprocess_text(user_input_text)
    if not cleaned_user_input:
        print("預處理後的使用者輸入為空，無法生成嵌入。")
        return pd.DataFrame()

    # 2. 為使用者輸入生成嵌入
    try:
        user_embedding_response = genai.embed_content(
            model=embedding_model,
            contents=[cleaned_user_input], # 嵌入 API 期望內容是列表
            task_type="SEMANTIC_SIMILARITY"
        )
        user_embedding = np.array(user_embedding_response.embeddings[0].values)
        if user_embedding.ndim == 1:
            user_embedding = user_embedding.reshape(1, -1) # 重塑為 2D 陣列以便於 cosine_similarity

    except Exception as e:
        print(f"為使用者輸入生成嵌入時發生錯誤: {e}")
        return pd.DataFrame()

    # 3. 計算餘弦相似度
    # cosine_similarity 函數期望兩個 2D 陣列
    similarities = cosine_similarity(user_embedding, embeddings_matrix).flatten()

    # 4. 建立相似度 Series 並排序
    similarity_scores = pd.Series(similarities, index=dataframe_with_embeddings.index)

    # 5. 獲取相似度最高的 Top N 歌曲
    # .nlargest 會自動排除輸入本身，因為輸入的嵌入不會在 embeddings_matrix 中找到與自己完全相同的（ unless the user input is *exactly* one of the songs)
    # 但為了健壯性，我們可以確保推薦的是不同的歌曲。
    # 由於這裡的使用者輸入是新的，所以無需排除自己。
    top_similar_indices = similarity_scores.nlargest(top_n).index

    # 6. 檢索推薦歌曲的詳細資訊
    recommendations = dataframe_with_embeddings.loc[top_similar_indices, ['Song Name', 'Artist']].copy()
    recommendations['Similarity_Score'] = similarity_scores.loc[top_similar_indices]

    return recommendations.reset_index(drop=True)

print("`get_recommendations_from_text` 函數已定義。")
```

```python
def run_recommender_prototype():
    print("\n--- 歡迎使用歌曲推薦系統原型！ ---")
    print("輸入您喜歡的歌曲類型、主題或情感描述，我將為您推薦類似的歌曲。")
    print("輸入 'exit' 或 'quit' 退出。")

    while True:
        user_input = input("\n請輸入您喜歡的歌曲描述 (例如: '關於愛與失落的悲傷歌曲', '充滿活力的夏日派對音樂'): \n> ")

        if user_input.lower() in ['exit', 'quit']:
            print("感謝您的使用，再見！")
            break

        if not user_input.strip():
            print("您的輸入為空，請輸入一些描述。")
            continue

        # 獲取推薦
        # 我們使用之前預處理好的 df_processed_with_embeddings 和 song_embeddings_matrix
        recommended_songs = get_recommendations_from_text(
            user_input_text=user_input,
            dataframe_with_embeddings=df_processed_with_embeddings,
            embeddings_matrix=song_embeddings_matrix,
            top_n=5 # 推薦前 5 首
        )

        if not recommended_songs.empty:
            print("\n--- 為您推薦的歌曲: ---")
            print(recommended_songs.to_string(index=False)) # 使用 to_string 以便更好地格式化輸出
        else:
            print("\n抱歉，未能根據您的描述找到相關推薦。請嘗試其他描述。")

# 運行主程式
if __name__ == "__main__":
    # 檢查是否成功載入了足夠的資料並生成了嵌入
    if 'df_processed_with_embeddings' in locals() and not df_processed_with_embeddings.empty:
        run_recommender_prototype()
    else:
        print("\n系統初始化失敗，請檢查之前的步驟是否成功載入資料和生成嵌入。")
```
